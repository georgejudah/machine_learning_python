```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# EXAMPLE 1 - A simple cost function

$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
#print(x_1)
```

```python
# Plot
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope & Derivatives
Challenge: Create a python function for the derivative of f(x) called df(x)

```python
def df(x):
    # Power rule
    return 2*x + 1
```

```python
# Plot function and derivate side by side
# figsize(float, float): These parameter are the width, height in inches.
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
# 1 rows 2 column - for stacking one below the other in the index 1
plt.subplot(1, 2, 1)

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color= 'blue', linewidth=3)

# 2 chart: Derivative
plt.subplot(1,2,2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5 )

plt.show()
```
